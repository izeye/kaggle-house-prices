```python
from sklearn.linear_model import LinearRegression
from math import sqrt

linear_regression = LinearRegression()
linear_regression.fit(X_train, Y_train)
linear_regression_mse = np.mean((linear_regression.predict(X_train) - Y_train) ** 2)
print(linear_regression_mse)
linear_regression_score = linear_regression.score(X_train, Y_train)
linear_regression_score
```

```python
coeff_df = pd.DataFrame(X_train.columns)
coeff_df.columns = ['Feature']
coeff_df['Correlation'] = pd.Series(linear_regression.coef_)
coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
models = pd.DataFrame({
    'Model': [
        linear_regression
    ],
    'Score': [
        linear_regression_score
    ]
})
models.sort_values(by='Score', ascending=False)
```

```python
Y_pred = models['Model'][0].predict(X_test)

submission = pd.DataFrame({
    "Id": test_df["Id"],
    "SalePrice": Y_pred
})
submission.loc[submission["SalePrice"] < 0, "SalePrice"] = -submission["SalePrice"]
submission.to_csv("../output/submission.csv", index=False)
```
